## 1.1. Скачать данные в закрепе к уроку

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

* id - уникальный ID дома

* date - дата продажи дома 

* price - стоимость продажи дома

* bedrooms - кол-во спален

* bathrooms - кол-во ванных комнат

* sqft_living - жилая площадь

* sqft_lot - общая площадь

* floors - кол-во этажей

* waterfront - набержная

* view - оченка вида от 0 до 4 

* condition - оценка состояния от 1 до 5

* grade - оценка от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна

* sqft_above - площадь дома, находящаяся выше земли

* sqft_basement - площадь дома, находящаяся ниже земли

* yr_built - год постройки дома

* yr_renovated - год ремонта дома

* zipcode - Индекс

* lat - широта

* long - долгота

* sqft_living15 - жилая площадь у 15 соседей

* sqft_lot15 - общая площадь у 15 соседей

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(9, 7))
plt.hist(df['price'])
plt.title('График стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

##### Около 70% приходится на самую дешёвую недвижимость

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(9, 7))
sns.histplot(df['sqft_living'], bins=20)
plt.title('График жилой площади')
plt.xlabel('Жилая площадь кв/м')
plt.ylabel('Количество');
```

##### Больше всего жилья с жилой площадью в 2000 кв/м

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(7, 9))
sns.histplot(df['yr_built'], bins=50)
plt.title('График года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

Отчетливо заметен строительный бум в 1960х, 1980х и 2000 годах. Кризис строительства во временна депрессии 1930х

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(9, 7))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
plt.xlabel('Цена')
plt.ylabel('Жилая площадь');
```

##### Жилая площадь дома тем больше, чем выше его стоимость

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
plt.xlabel('Цена')
plt.ylabel('Оценка');
```

##### Стоимость дома выше, чем выше качество конструкции и дизайна

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка вида')
plt.title('График цены по оценке вида');
```

##### Наиболее дорогие дома имеют оценку вида со значением 4

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Оценка')
plt.title('График цены от оценки');
```

##### Дома с оценкой конструкции и дизайна 13 дороже остальных

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Этажи')
plt.title('График цен по этажам');
```

##### Дороже всех дома с 2 этажами + подвал
